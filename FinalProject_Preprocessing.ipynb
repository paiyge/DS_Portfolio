```python
import os
import sys
import shutil
import random
from pathlib import Path
from glob import glob
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

```python
path = Path(os.getcwd())
path = str(path)
print(path)
sys.path.insert(1, path)
```

```python
# Define the path to the folder containing subfolders for each artist
data_dir = path + "/Final Project Data"
print(data_dir)
```

```python
# Define the paths for train, test, and validation directories
train_dir = path + "/train"
test_dir = path + "/test"
valid_dir = path + "/validation"
```

```python
# Define the ratio for train-test split and train-validation split
test_ratio = 0.2
valid_ratio = 0.1
```

```python
# Create train, test, and validation directories
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
os.makedirs(valid_dir, exist_ok=True)
```

```python
# Iterate through each subfolder (artist) in the data directory
for artist_folder in os.listdir(data_dir):
    artist_path = os.path.join(data_dir, artist_folder)
    
    # Skip if it's not a directory
    if not os.path.isdir(artist_path):
        continue
    
    # Get list of image filenames in the artist's folder
    images = os.listdir(artist_path)
    
    # Print all filenames in the artist's folder
    print(f"Files in {artist_folder}:")
    for image in images:
        print(image)
    
    # Check if there are images in the folder
    if len(images) == 0:
        print(f"No images found in {artist_folder}. Skipping...")
        continue
    
    # Split the images into train and test sets
    train_images, test_images = train_test_split(images, test_size=test_ratio, random_state=42)
    
    # Further split the train set into train and validation sets
    train_images, valid_images = train_test_split(train_images, test_size=valid_ratio, random_state=42)
    
    # Copy images to respective directories
    for image in train_images:
        src = os.path.join(artist_path, image)
        dst = os.path.join(train_dir, artist_folder, image)
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        shutil.copy(src, dst)
        
    for image in test_images:
        src = os.path.join(artist_path, image)
        dst = os.path.join(test_dir, artist_folder, image)
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        shutil.copy(src, dst)
        
    for image in valid_images:
        src = os.path.join(artist_path, image)
        dst = os.path.join(valid_dir, artist_folder, image)
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        shutil.copy(src, dst)
```
